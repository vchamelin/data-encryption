```python
def lfsr(seed, taps, length):
    sr, xor = [int(i) for i in seed], 0
    bits = []
    for _ in range(length):
        xor = sr[len(sr)-1]
        for t in taps:
            sr[t-1] = (xor + int(sr[t-1])) % 2 
        bits.append(xor)
        sr = sr[:-1]
        sr.insert(0, xor)
    return bits

# x^18 + x^7 + 1
seed = '100100011010001010'
taps = [7]
length = 6000

sequence = lfsr(seed, taps, length)
print(sequence)
```

```python
print(f"Частотний тест! \nResult: {sequence.count(1) / length}")
```

```python
count_one_diff = sum(sequence[i] ^ sequence[i+1] for i in range(length - 1))
result = count_one_diff / (length - 1)

print(f"Диференційний тест! \nResult: {result}\n")
```

```python
def get_all_possible_bitarrays(length: int):
    result = {}
    for i in range(2**length):
        a = [int(x) for x in bin(i)[2:]]
        a = [0]*(length - len(a)) + a
        result[''.join(map(str, a))] = 0
    return result

rang_test_length = 4
rang_test = get_all_possible_bitarrays(rang_test_length)


for i in range(rang_test_length, length):
    rang_test[''.join(map(str, sequence[i-rang_test_length:i]))] += 1

print("Ранговий тест:")

for key, value in rang_test.items():
    print(f"{key} -> {value}")
print('\n')
```

```python
non_linear_complexity = 1
while True:
    non_linear_test = {}
    for i in range(non_linear_complexity, length):
        sequence_key = ''.join(map(str, sequence[i-non_linear_complexity:i]))
        if sequence_key in non_linear_test and non_linear_test[sequence_key] != sequence[i]:
            break
        non_linear_test[sequence_key] = sequence[i]
    else: 
        break
    non_linear_complexity += 1

print(f"Нелінійна складність: {non_linear_complexity}")
```
